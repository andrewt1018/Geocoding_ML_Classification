```python
# Necessary imports
import pandas as pd
import numpy as np
import pickle
```

```python
# Load in all of the necessary data
f = open("all_data.pkl", "rb")
df = pickle.load(f)
f.close()

# Randomize training data
def get_training(df):
    df_shuffled = df.sample(frac=1)
    valid = df_shuffled.loc[df_shuffled['expected'] == 0].loc[df_shuffled['distances'] != 0]
    other_valid = df_shuffled.loc[df_shuffled['expected'] == 0].loc[df_shuffled['distances'] == 0]
    valid = pd.concat([valid, other_valid])
    invalid = df_shuffled.loc[df_shuffled['expected'] == 1]
    return pd.concat([valid.iloc[0:len(invalid)], invalid])

def randomize(X_train, y_train):
    df = X_train.copy()
    df['labels'] = y_train
    df = df.sample(frac=1)
    return (df.iloc[:,:-1], df.iloc[:,-1])

df_ready = get_training(df)
df_ready = df_ready.iloc[:, 1:]  # Drop the index column
df_ready['expected'].value_counts()
```

```python
# Separating data into training data and testing data
features = df_ready.drop(['expected'], axis=1, inplace=False)
labels = df_ready['expected']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, labels, 
                                                    shuffle = True, 
                                                    test_size = 0.2, 
                                                    random_state = 1)
# Show the Training and Testing Data
print('Shape of training features:', X_train.shape)
print('Shape of testing features:', X_test.shape)
print('Shape of training labels:', y_train.shape)
print('Shape of training labels:', y_test.shape)
```

```python
from decision_tree import DecisionTree

tree = DecisionTree(2, 10)
X_train, y_train = tree.shuffle(X_train, y_train)
tree.fit(X_train, y_train, cost_func='gini')
```

```python
from random_forest import RandomForest

forest = RandomForest(max_depth=5, num_trees=10)
X_train, y_train = randomize(X_train, y_train)
forest.fit(X_train, y_train, 'gini')
forest.evaluate(X_test, y_test)
```
